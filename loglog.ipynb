```python

def plotData(ret, label=""):
    out = np.bincount(ret["nodeRetweets"])
    plt.loglog(np.arange(len(out)), out, label=label+" retweets")
    out = np.bincount(ret["nodeTweets"])
    plt.loglog(np.arange(len(out)), out, label=label+" tweets")
    
fig = plt.figure()
fig.set_size_inches(15,8)
plotData(ret_big, "Monday")
plotData(ret_med, "Wednesday")
plt.legend()
```
